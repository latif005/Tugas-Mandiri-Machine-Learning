```python
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
# Data dari soal
data = {
    'Sebenarnya': ['Lulus', 'Lulus', 'Lulus', 'Lulus', 'Lulus',
                         'Tidak Lulus', 'Tidak Lulus', 'Tidak Lulus', 'Tidak Lulus', 'Tidak Lulus'],
    'Prediksi': ['Lulus', 'Lulus', 'Lulus', 'Tidak Lulus', 'Tidak Lulus',
                       'Lulus', 'Tidak Lulus', 'Tidak Lulus', 'Tidak Lulus', 'Tidak Lulus']
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
# Ubah label
df['Actual'] = df['Sebenarnya'].map({'Lulus':1, 'Tidak Lulus':0})
df['Pred'] = df['Prediksi'].map({'Lulus':1, 'Tidak Lulus':0})
```

```python
# Confusion Matrix
cm = confusion_matrix(df['Actual'], df['Pred'])
print("Confusion Matrix:\n", cm)
```

```python
# Metrics
acc = accuracy_score(df['Actual'], df['Pred'])
prec = precision_score(df['Actual'], df['Pred'])
rec = recall_score(df['Actual'], df['Pred'])

print("\nAccuracy :", round(acc*100,2), "%")
print("Precision:", round(prec*100,2), "%")
print("Recall   :", round(rec*100,2), "%")
```
